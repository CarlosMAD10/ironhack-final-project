## 1. NASDAQ

```python
nasdaq_data = pd.read_csv("./data/nasdaq.csv")
```

```python
nasdaq_5 = nasdaq_data.sample(5)
```

```python
nasdaq_5
```

```python
nasdaq_5["price"] = nasdaq_5["Last Sale"].apply(lambda x: float(x.replace("$", "")))
```

```python
errors = {}
for symbol in nasdaq_5["Symbol"]:
    y, predicted_price, row = make_prediction(symbol, cluster_model, class_model, reg_model, pivot_date="2021-09-27")
    real_price = float(nasdaq_5.loc[nasdaq_5["Symbol"] == symbol,"price"])
    errors[symbol] = 100* (predicted_price - real_price)/real_price
```
